```python
import pandas as pd
import matplotlib.pyplot as plt

print("=" * 50)
print("🎬 NETFLIX MOVIE ANALYSIS")
print("=" * 50)

# 1. Basic stats
print(f"\nTotal titles: {len(df)}")
print(f"Movies: {len(df[df['type'] == 'Movie'])}")
print(f"TV Shows: {len(df[df['type'] == 'TV Show'])}")

# 2. Top genres
print("\n🎭 TOP 5 GENRES:")
df['genres'] = df['listed_in'].str.split(', ')
all_genres = df['genres'].explode()
print(all_genres.value_counts().head(5))

# 3. Movies by year
print("\n📅 TOP 5 YEARS WITH MOST RELEASES:")
print(df['release_year'].value_counts().head(5))

# 4. Top ratings
print("\n⭐ TOP 5 RATINGS:")
print(df['rating'].value_counts().head(5))

# Charts
plt.figure(figsize=(10,5))
df['release_year'].value_counts().head(10).sort_index().plot(kind='bar')
plt.title('Top 10 Years with Most Releases')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()

plt.figure(figsize=(8,5))
all_genres.value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Genres')
plt.xlabel('Number of Titles')
plt.show()
```
